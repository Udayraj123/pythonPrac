```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

arr = []
num_arrs=5
sizes=(20,60)
max_size=0
for i in range(num_arrs):
    arr_size = np.random.randint(sizes[0],sizes[1])
    arr.append(np.random.randn(arr_size))
    max_size=max(arr_size,max_size)

plt.figure('Single Plot',figsize=(20,5))
plt.xlabel('time')
plt.ylabel('speed')
plt.title('speed at timepoint t')
plt.grid(True)
for i in arr:
    plt.figure('Single Plot',figsize=(20,5))
    plt.plot(i)
```
